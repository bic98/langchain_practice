````python
async def discover_tools():
    """
    MCP ì„œë²„ì— ì—°ê²°í•˜ì—¬ ì‚¬ìš© ê°€ëŠ¥í•œ ë„êµ¬ë¥¼ ë°œê²¬í•©ë‹ˆë‹¤.
    
    ì™œ async í•¨ìˆ˜ë¥¼ ì‚¬ìš©í• ê¹Œ?
    - MCPëŠ” ë¹„ë™ê¸° í”„ë¡œí† ì½œì…ë‹ˆë‹¤
    - I/O ì‘ì—…(ë„¤íŠ¸ì›Œí¬, íŒŒì¼)ì„ íš¨ìœ¨ì ìœ¼ë¡œ ì²˜ë¦¬í•©ë‹ˆë‹¤
    - ì—¬ëŸ¬ ì‘ì—…ì„ ë™ì‹œì— ì²˜ë¦¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤
    
    Returns:
        ì‚¬ìš© ê°€ëŠ¥í•œ ë„êµ¬ë“¤ì˜ ì •ë³´ ë¦¬ìŠ¤íŠ¸
    """
    # ==================== ìƒ‰ìƒ ì½”ë“œ (ë¡œê·¸ ê°€ë…ì„±) ====================
    BLUE = "\033[94m"
    GREEN = "\033[92m"
    RESET = "\033[0m"
    SEP = "=" * 40
    
    # ==================== 1. ì„œë²„ íŒŒë¼ë¯¸í„° ì„¤ì • ====================
    # StdioServerParameters: stdioë¥¼ í†µí•´ ì„œë²„ì™€ í†µì‹ í•˜ê¸° ìœ„í•œ ì„¤ì •
    server_params = StdioServerParameters(
        command="python",  # ì„œë²„ ì‹¤í–‰ ëª…ë ¹ì–´
        args=[mcp_server_path],  # ì„œë²„ ìŠ¤í¬ë¦½íŠ¸ ê²½ë¡œ
    )
    
    print(f"{BLUE}{SEP}")
    print("ğŸ” DISCOVERY PHASE: MCP ì„œë²„ ì—°ê²° ì¤‘...")
    print(f"{SEP}{RESET}")
    
    # ==================== 2. stdio í´ë¼ì´ì–¸íŠ¸ë¡œ ì—°ê²° ====================
    # stdio_client: ì„œë²„ í”„ë¡œì„¸ìŠ¤ë¥¼ ì‹œì‘í•˜ê³  stdin/stdout ìŠ¤íŠ¸ë¦¼ ìƒì„±
    # async with: ë¹„ë™ê¸° ì»¨í…ìŠ¤íŠ¸ ë§¤ë‹ˆì € (ìë™ìœ¼ë¡œ ë¦¬ì†ŒìŠ¤ ì •ë¦¬)
    async with stdio_client(server_params) as (read, write):
        # ==================== 3. í´ë¼ì´ì–¸íŠ¸ ì„¸ì…˜ ìƒì„± ====================
        # ClientSession: MCP í”„ë¡œí† ì½œ í†µì‹ ì„ ê´€ë¦¬
        async with ClientSession(read, write) as session:
            # ==================== 4. ì„¸ì…˜ ì´ˆê¸°í™” ====================
            # initialize(): MCP í•¸ë“œì…°ì´í¬ ìˆ˜í–‰
            # - í”„ë¡œí† ì½œ ë²„ì „ í˜‘ìƒ
            # - ì„œë²„ capabilities í™•ì¸
            print(f"{BLUE}ğŸ“¡ MCP ì—°ê²° ì´ˆê¸°í™” ì¤‘...{RESET}")
            await session.initialize()
            
            # ==================== 5. ë„êµ¬ ëª©ë¡ ê°€ì ¸ì˜¤ê¸° ====================
            # list_tools(): ì„œë²„ê°€ ì œê³µí•˜ëŠ” ëª¨ë“  ë„êµ¬ ì¡°íšŒ
            print(f"{BLUE}ğŸ” ì‚¬ìš© ê°€ëŠ¥í•œ ë„êµ¬ ê²€ìƒ‰ ì¤‘...{RESET}")
            tools = await session.list_tools()
            
            # ==================== 6. ë„êµ¬ ì •ë³´ í¬ë§·íŒ… ====================
            # ë„êµ¬ ì •ë³´ë¥¼ ì‚¬ìš©í•˜ê¸° ì‰¬ìš´ í˜•íƒœë¡œ ë³€í™˜
            tool_info = []
            for tool_type, tool_list in tools:
                if tool_type == "tools":
                    for tool in tool_list:
                        tool_info.append({
                            "name": tool.name,  # ë„êµ¬ ì´ë¦„
                            "description": tool.description,  # ì„¤ëª…
                            "schema": tool.inputSchema  # ì…ë ¥ ìŠ¤í‚¤ë§ˆ
                        })
            
            print(f"{GREEN}âœ… {len(tool_info)}ê°œì˜ ë„êµ¬ ë°œê²¬ ì™„ë£Œ{RESET}")
            print(f"{SEP}")
            return tool_info

print("âœ… ë„êµ¬ ë°œê²¬ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ")
````

### Clientì˜ í•µì‹¬ ì—­í• : ë„êµ¬ ì‹¤í–‰

ì´ì œ ë°œê²¬í•œ ë„êµ¬ë¥¼ ì‹¤ì œë¡œ ì‹¤í–‰í•˜ëŠ” Clientë¥¼ ë§Œë“¤ê² ìŠµë‹ˆë‹¤.

### ì‹¤í–‰ í”„ë¡œì„¸ìŠ¤

```
1. ë„êµ¬ ì´ë¦„ê³¼ ì¸ì ë°›ê¸°
   â†“
2. ì„œë²„ ì—°ê²° ìƒì„±
   â†“

   â†“
4. ì„œë²„ê°€ ë„êµ¬ ì‹¤í–‰
   â†“
```

````python
async def execute_tool(tool_name: str, arguments: Dict[str, Any]):
    """
    MCP ì„œë²„ì˜ íŠ¹ì • ë„êµ¬ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤.
    
    ì™œ ë§¤ë²ˆ ìƒˆë¡œìš´ ì—°ê²°ì„ ë§Œë“¤ê¹Œ?
    - ê° ë„êµ¬ í˜¸ì¶œì„ ë…ë¦½ì ìœ¼ë¡œ ì²˜ë¦¬í•©ë‹ˆë‹¤
    - ìƒíƒœ ë¬¸ì œë¥¼ ë°©ì§€í•©ë‹ˆë‹¤
    - ë””ë²„ê¹…ì´ ì‰½ìŠµë‹ˆë‹¤
    
    í”„ë¡œë•ì…˜ì—ì„œëŠ”?
    - ì—°ê²°ì„ ì¬ì‚¬ìš©í•˜ì—¬ ì„±ëŠ¥ì„ í–¥ìƒì‹œí‚¬ ìˆ˜ ìˆìŠµë‹ˆë‹¤
    - í•˜ì§€ë§Œ í•™ìŠµìš©ìœ¼ë¡œëŠ” ì´ ë°©ì‹ì´ ë” ëª…í™•í•©ë‹ˆë‹¤
    
    Args:
        tool_name: ì‹¤í–‰í•  ë„êµ¬ì˜ ì´ë¦„
        arguments: ë„êµ¬ì— ì „ë‹¬í•  ì¸ì (ë”•ì…”ë„ˆë¦¬)
        
    Returns:
        ë„êµ¬ ì‹¤í–‰ ê²°ê³¼
    """
    # ==================== ìƒ‰ìƒ ì½”ë“œ ====================
    BLUE = "\033[94m"
    GREEN = "\033[92m"
    YELLOW = "\033[93m"
    RESET = "\033[0m"
    SEP = "-" * 40
    
    # ==================== ì„œë²„ íŒŒë¼ë¯¸í„° ì„¤ì • ====================
    server_params = StdioServerParameters(
        command="python",
        args=[mcp_server_path],
    )
    
    print(f"{YELLOW}{SEP}")
    print(f"âš™ï¸ EXECUTION PHASE: '{tool_name}' ë„êµ¬ ì‹¤í–‰ ì¤‘")
    print(f"ğŸ“‹ ì¸ì: {json.dumps(arguments, indent=2, ensure_ascii=False)}")
    print(f"{SEP}{RESET}")
    
    # ==================== ì„œë²„ ì—°ê²° ë° ì‹¤í–‰ ====================
    async with stdio_client(server_params) as (read, write):
        async with ClientSession(read, write) as session:
            # ì„¸ì…˜ ì´ˆê¸°í™”
            await session.initialize()
            
            # ==================== ë„êµ¬ í˜¸ì¶œ ====================
            # call_tool(): ì„œë²„ì˜ íŠ¹ì • ë„êµ¬ë¥¼ ì‹¤í–‰
            # - tool_name: ì‹¤í–‰í•  ë„êµ¬ì˜ ì´ë¦„
            # - arguments: ë„êµ¬ì— ì „ë‹¬í•  ì¸ì
            print(f"{BLUE}ğŸ“¡ MCP ì„œë²„ë¡œ ìš”ì²­ ì „ì†¡ ì¤‘...{RESET}")
            result = await session.call_tool(tool_name, arguments)
            
            print(f"{GREEN}âœ… ë„êµ¬ ì‹¤í–‰ ì™„ë£Œ{RESET}")
            
            # ê²°ê³¼ ë¯¸ë¦¬ë³´ê¸° (ë„ˆë¬´ ê¸¸ë©´ ìë¥´ê¸°)
            result_preview = str(result)
            if len(result_preview) > 150:
                result_preview = result_preview[:147] + "..."
                
            print(f"{BLUE}ğŸ“Š ê²°ê³¼: {result_preview}{RESET}")
            print(f"{SEP}")
            
            return result

print("âœ… ë„êµ¬ ì‹¤í–‰ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ")
````


````python
async def query_claude(prompt: str, tool_info: List[Dict], previous_messages=None):
    """
    Claudeì—ê²Œ ì§ˆë¬¸ì„ ë³´ë‚´ê³  í•„ìš”ì‹œ ë„êµ¬ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤.
    
    ì „ì²´ í”„ë¡œì„¸ìŠ¤:
    1. ë„êµ¬ ì •ë³´ë¥¼ ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ì— í¬í•¨
    2. Claudeì—ê²Œ ì§ˆë¬¸ ì „ì†¡
    3. Claudeê°€ ë„êµ¬ ì‚¬ìš©ì„ ìš”ì²­í•˜ë©´ ì‹¤í–‰
    4. ê²°ê³¼ë¥¼ Claudeì—ê²Œ ì „ë‹¬í•˜ì—¬ í•´ì„
    5. ìµœì¢… ë‹µë³€ ë°˜í™˜
    
    Args:
        prompt: ì‚¬ìš©ìì˜ ì§ˆë¬¸
        tool_info: ì‚¬ìš© ê°€ëŠ¥í•œ ë„êµ¬ ì •ë³´
        previous_messages: ì´ì „ ëŒ€í™” ë‚´ì—­ (ì»¨í…ìŠ¤íŠ¸ ìœ ì§€ìš©)
        
    Returns:
        (Claudeì˜ ìµœì¢… ë‹µë³€, ì—…ë°ì´íŠ¸ëœ ë©”ì‹œì§€ ë¦¬ìŠ¤íŠ¸)
    """
    # ==================== ìƒ‰ìƒ ì½”ë“œ ====================
    PURPLE = "\033[95m"
    BLUE = "\033[94m"
    GREEN = "\033[92m"
    YELLOW = "\033[93m"
    RESET = "\033[0m"
    SEP = "=" * 40
    
    if previous_messages is None:
        previous_messages = []
    
    print(f"{PURPLE}{SEP}")
    print("ğŸ§  REASONING PHASE: Claudeë¡œ ì§ˆë¬¸ ì²˜ë¦¬ ì¤‘")
    print(f"ğŸ”¤ ì§ˆë¬¸: \"{prompt}\"")
    print(f"{SEP}{RESET}")
    
    # ==================== 1. ë„êµ¬ ì„¤ëª… í¬ë§·íŒ… ====================
    # Claudeê°€ ì´í•´í•  ìˆ˜ ìˆëŠ” í˜•ì‹ìœ¼ë¡œ ë„êµ¬ ì •ë³´ ë³€í™˜
    tool_descriptions = "\n\n".join([
        f"ë„êµ¬: {tool['name']}\n"
        f"ì„¤ëª…: {tool['description']}\n"
        f"ìŠ¤í‚¤ë§ˆ: {json.dumps(tool['schema'], indent=2, ensure_ascii=False)}"
        for tool in tool_info
    ])
    
    # ==================== 2. ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ êµ¬ì„± ====================
    # Claudeì—ê²Œ ë„êµ¬ ì‚¬ìš© ë°©ë²•ì„ ì•Œë ¤ì¤ë‹ˆë‹¤
    system_prompt = f"""ë‹¹ì‹ ì€ MCP(Model Context Protocol)ë¥¼ í†µí•´ íŠ¹í™”ëœ ë„êµ¬ì— ì ‘ê·¼í•  ìˆ˜ ìˆëŠ” AI ì–´ì‹œìŠ¤í„´íŠ¸ì…ë‹ˆë‹¤.
    
ì‚¬ìš© ê°€ëŠ¥í•œ ë„êµ¬:
{tool_descriptions}

ë„êµ¬ë¥¼ ì‚¬ìš©í•´ì•¼ í•  ë•ŒëŠ” ë‹¤ìŒ JSON í˜•ì‹ìœ¼ë¡œ ì‘ë‹µí•˜ì„¸ìš”:
{{
    "tool": "ë„êµ¬_ì´ë¦„",
    "arguments": {{
        "ì¸ì1": "ê°’1",
        "ì¸ì2": "ê°’2"
    }}
}}

ë„êµ¬ë¥¼ ì‚¬ìš©í•  ë•ŒëŠ” JSONë§Œ ì¶œë ¥í•˜ê³  ë‹¤ë¥¸ í…ìŠ¤íŠ¸ëŠ” í¬í•¨í•˜ì§€ ë§ˆì„¸ìš”.
ì¼ë°˜ ì‘ë‹µì¼ ë•ŒëŠ” ìì—°ìŠ¤ëŸ½ê²Œ ëŒ€ë‹µí•˜ì„¸ìš”.
"""
    
    # ==================== 3. ë©”ì‹œì§€ êµ¬ì„± ====================
    # ì‹œìŠ¤í…œ ë©”ì‹œì§€ëŠ” ì œì™¸í•˜ê³  user/assistant ë©”ì‹œì§€ë§Œ í¬í•¨
    filtered_messages = [
        msg for msg in previous_messages 
        if msg["role"] != "system"
    ]
    
    messages = filtered_messages.copy()
    messages.append({"role": "user", "content": prompt})
    
    # ==================== 4. Claude API í˜¸ì¶œ ====================
    print(f"{BLUE}ğŸ“¡ Claude APIë¡œ ìš”ì²­ ì „ì†¡ ì¤‘...{RESET}")
    
    response = client.messages.create(
        model="claude-3-5-sonnet-20240620",
        max_tokens=4000,
        system=system_prompt,  # ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ (ë³„ë„ íŒŒë¼ë¯¸í„°)
        messages=messages  # ëŒ€í™” ë‚´ì—­
    )
    
    claude_response = response.content[0].text
    print(f"{GREEN}âœ… Claude ì‘ë‹µ ìˆ˜ì‹ {RESET}")
    
    # ==================== 5. ë„êµ¬ ì‚¬ìš© ê°ì§€ ====================
    # Claudeì˜ ì‘ë‹µì—ì„œ JSON íŒ¨í„´ ì°¾ê¸°
    import re
    json_match = re.search(r'(\{[\s\S]*\})', claude_response)
    
    if json_match:
        try:
            json_str = json_match.group(1)
            print(f"{YELLOW}ğŸ” ë„êµ¬ ì‚¬ìš© ê°ì§€{RESET}")
            print(f"{BLUE}ğŸ“¦ ì¶”ì¶œëœ JSON: {json_str}{RESET}")
            
            tool_request = json.loads(json_str)
            
            if "tool" in tool_request and "arguments" in tool_request:
                tool_name = tool_request["tool"]
                arguments = tool_request["arguments"]
                
                print(f"{YELLOW}ğŸ”§ Claudeê°€ ë„êµ¬ ì‚¬ìš© ìš”ì²­: {tool_name}{RESET}")
                
                # ==================== 6. ë„êµ¬ ì‹¤í–‰ ====================
                tool_result = await execute_tool(tool_name, arguments)
                
                # ê²°ê³¼ë¥¼ ë¬¸ìì—´ë¡œ ë³€í™˜
                if not isinstance(tool_result, str):
                    tool_result = str(tool_result)
                
                # ë©”ì‹œì§€ íˆìŠ¤í† ë¦¬ ì—…ë°ì´íŠ¸
                messages.append({"role": "assistant", "content": claude_response})
                messages.append({"role": "user", "content": f"ë„êµ¬ ì‹¤í–‰ ê²°ê³¼: {tool_result}"})
                
                print(f"{PURPLE}ğŸ”„ Claudeì—ê²Œ ê²°ê³¼ í•´ì„ ìš”ì²­ ì¤‘...{RESET}")
                
                # ==================== 7. ê²°ê³¼ í•´ì„ ====================
                # Claudeì—ê²Œ ë„êµ¬ ê²°ê³¼ë¥¼ ìì—°ì–´ë¡œ ì„¤ëª…í•˜ë„ë¡ ìš”ì²­
                final_response = client.messages.create(
                    model="claude-3-5-sonnet-20240620",
                    max_tokens=4000,
                    system=system_prompt,
                    messages=messages
                )
                
                print(f"{GREEN}âœ… ìµœì¢… ì‘ë‹µ ì¤€ë¹„ ì™„ë£Œ{RESET}")
                print(f"{SEP}")
                
                return final_response.content[0].text, messages
                
        except (json.JSONDecodeError, KeyError, AttributeError) as e:
            print(f"{YELLOW}âš ï¸ ë„êµ¬ ì‚¬ìš© ê°ì§€ ì‹¤íŒ¨: {str(e)}{RESET}")
    
    print(f"{GREEN}âœ… ì‘ë‹µ ì¤€ë¹„ ì™„ë£Œ{RESET}")
    print(f"{SEP}")
    
    return claude_response, messages

print("âœ… Claude ì¿¼ë¦¬ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ")
````

### ğŸ’¡ ì´í•´í•˜ê¸°: JSON íŒ¨í„´ ë§¤ì¹­
