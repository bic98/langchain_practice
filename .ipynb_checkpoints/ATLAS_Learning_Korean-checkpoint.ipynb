````python
class DataManager:
    """
    ë°ì´í„° ê´€ë¦¬ í´ë˜ìŠ¤
    
    í•™ìƒ í”„ë¡œí•„, ìº˜ë¦°ë”, í•  ì¼ ëª©ë¡ì„ ì¤‘ì•™ì—ì„œ ê´€ë¦¬í•©ë‹ˆë‹¤.
    ëª¨ë“  ì—ì´ì „íŠ¸ëŠ” ì´ í´ë˜ìŠ¤ë¥¼ í†µí•´ ë°ì´í„°ì— ì ‘ê·¼í•©ë‹ˆë‹¤.
    """
    
    def __init__(self):
        """ë°ì´í„° ì €ì¥ì†Œ ì´ˆê¸°í™” (ëª¨ë‘ Noneìœ¼ë¡œ ì‹œì‘)"""
        self.profile_data = None
        self.calendar_data = None
        self.task_data = None
    
    def load_data(self, profile_json: str, calendar_json: str, task_json: str):
        """
        JSON ë¬¸ìì—´ì„ íŒŒì‹±í•˜ì—¬ ë°ì´í„° ë¡œë“œ
        
        Args:
            profile_json: í•™ìƒ í”„ë¡œí•„ JSON
            calendar_json: ìº˜ë¦°ë” ì´ë²¤íŠ¸ JSON
            task_json: í•  ì¼ ëª©ë¡ JSON
        """
        self.profile_data = json.loads(profile_json)
        self.calendar_data = json.loads(calendar_json)
        self.task_data = json.loads(task_json)
    
    def get_student_profile(self, student_id: str) -> Dict:
        """
        í•™ìƒ IDë¡œ í”„ë¡œí•„ ê²€ìƒ‰
        
        Args:
            student_id: í•™ìƒ ê³ ìœ  ID
        
        Returns:
            Dict: í•™ìƒ í”„ë¡œí•„ (ì—†ìœ¼ë©´ None)
        
        êµ¬í˜„ ë°©ì‹:
            - generator expression + next()ë¡œ íš¨ìœ¨ì  ê²€ìƒ‰
            - ì „ì²´ ë¦¬ìŠ¤íŠ¸ ìˆœíšŒë¥¼ í”¼í•¨
        """
        if self.profile_data:
            return next((p for p in self.profile_data["profiles"]
                        if p["id"] == student_id), None)
        return None
    
    def parse_datetime(self, dt_str: str) -> datetime:
        """
        ìŠ¤ë§ˆíŠ¸ ë‚ ì§œ/ì‹œê°„ íŒŒì‹± (íƒ€ì„ì¡´ ì²˜ë¦¬ í¬í•¨)
        
        Args:
            dt_str: ISO í˜•ì‹ ë‚ ì§œ/ì‹œê°„ ë¬¸ìì—´
        
        Returns:
            datetime: UTC íƒ€ì„ì¡´ìœ¼ë¡œ ë³€í™˜ëœ datetime ê°ì²´
        
        ì²˜ë¦¬ ë°©ì‹:
            1. íƒ€ì„ì¡´ ì •ë³´ê°€ ìˆìœ¼ë©´ íŒŒì‹± í›„ UTCë¡œ ë³€í™˜
            2. íƒ€ì„ì¡´ ì •ë³´ê°€ ì—†ìœ¼ë©´ UTCë¡œ ê°€ì •
        """
        try:
            # 'Z'ë¥¼ '+00:00'ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ íŒŒì‹±
            dt = datetime.fromisoformat(dt_str.replace('Z', '+00:00'))
            return dt.astimezone(timezone.utc)
        except ValueError:
            # íƒ€ì„ì¡´ ì—†ìœ¼ë©´ UTCë¡œ ê°€ì •
            dt = datetime.fromisoformat(dt_str)
            return dt.replace(tzinfo=timezone.utc)
    
    def get_upcoming_events(self, days: int = 7) -> List[Dict]:
        """
        í–¥í›„ Nì¼ê°„ì˜ ì¼ì • ì¡°íšŒ
        
        Args:
            days: ì¡°íšŒí•  ê¸°ê°„ (ê¸°ë³¸ê°’: 7ì¼)
        
        Returns:
            List[Dict]: ì‹œê°„ìˆœìœ¼ë¡œ ì •ë ¬ëœ ì´ë²¤íŠ¸ ëª©ë¡
        
        êµ¬í˜„ ë¡œì§:
            - í˜„ì¬ ì‹œê°(UTC)ë¶€í„° daysì¼ í›„ê¹Œì§€ì˜ ì´ë²¤íŠ¸ë§Œ í•„í„°ë§
            - ì˜ëª»ëœ ë°ì´í„°ëŠ” warning ì¶œë ¥ í›„ ìŠ¤í‚µ
        """
        if not self.calendar_data:
            return []
        
        now = datetime.now(timezone.utc)
        future = now + timedelta(days=days)
        
        events = []
        for event in self.calendar_data.get("events", []):
            try:
                start_time = self.parse_datetime(event["start"]["dateTime"])
                
                # í˜„ì¬ ~ future ë²”ìœ„ì˜ ì´ë²¤íŠ¸ë§Œ í¬í•¨
                if now <= start_time <= future:
                    events.append(event)
            except (KeyError, ValueError) as e:
                print(f"âš ï¸  ì´ë²¤íŠ¸ ì²˜ë¦¬ ì˜¤ë¥˜: {str(e)}")
                continue
        
        return events
    
    def get_active_tasks(self) -> List[Dict]:
        """
        ë¯¸ì™„ë£Œ ê³¼ì œ ëª©ë¡ ì¡°íšŒ
        
        Returns:
            List[Dict]: í™œì„± ê³¼ì œ ëª©ë¡ (íŒŒì‹±ëœ ë§ˆê°ì¼ í¬í•¨)
        
        í•„í„°ë§ ì¡°ê±´:
            1. statusê°€ 'needsAction' (ë¯¸ì™„ë£Œ)
            2. ë§ˆê°ì¼ì´ í˜„ì¬ ì‹œê°ë³´ë‹¤ ë¯¸ë˜
        
        ì¶”ê°€ ê¸°ëŠ¥:
            - ê° taskì— 'due_datetime' í•„ë“œ ì¶”ê°€ (íŒŒì‹±ëœ datetime ê°ì²´)
        """
        if not self.task_data:
            return []
        
        now = datetime.now(timezone.utc)
        active_tasks = []
        
        for task in self.task_data.get("tasks", []):
            try:
                due_date = self.parse_datetime(task["due"])
                
                # ë¯¸ì™„ë£Œ + ë¯¸ë˜ ë§ˆê°ì¼ì¸ ê³¼ì œë§Œ í¬í•¨
                if task["status"] == "needsAction" and due_date > now:
                    # íŒŒì‹±ëœ datetime ì¶”ê°€ (ì—ì´ì „íŠ¸ê°€ ì‰½ê²Œ ì‚¬ìš© ê°€ëŠ¥)
                    task["due_datetime"] = due_date
                    active_tasks.append(task)
            except (KeyError, ValueError) as e:
                print(f"âš ï¸  ê³¼ì œ ì²˜ë¦¬ ì˜¤ë¥˜: {str(e)}")
                continue
        
        return active_tasks

print("âœ… DataManager í´ë˜ìŠ¤ ì •ì˜ ì™„ë£Œ")
````

````python
class ReActAgent:
    """
    ReACT íŒ¨í„´ ê¸°ë°˜ ì—ì´ì „íŠ¸ ë² ì´ìŠ¤ í´ë˜ìŠ¤
    
    ëª¨ë“  ì „ë¬¸ ì—ì´ì „íŠ¸(Planner, NoteWriter, Advisor)ì˜ ë¶€ëª¨ í´ë˜ìŠ¤ì…ë‹ˆë‹¤.
    ê³µí†µ ë„êµ¬ì™€ ê¸°ë³¸ ë¡œì§ì„ ì œê³µí•©ë‹ˆë‹¤.
    """
    
    def __init__(self, llm):
        """
        Args:
            llm: ì–¸ì–´ ëª¨ë¸ ì¸ìŠ¤í„´ìŠ¤
        """
        self.llm = llm
        
        # Few-shot í•™ìŠµ ì˜ˆì œ ì €ì¥ì†Œ (ë‚˜ì¤‘ì— ì‚¬ìš©)
        self.few_shot_examples = []
        
        # ì‚¬ìš© ê°€ëŠ¥í•œ ë„êµ¬ ë”•ì…”ë„ˆë¦¬
        self.tools = {
            "search_calendar": self.search_calendar,
            "analyze_tasks": self.analyze_tasks,
            "check_learning_style": self.check_learning_style,
            "check_performance": self.check_performance
        }
    
    async def search_calendar(self, state: AcademicState) -> List[Dict]:
        """
        ìº˜ë¦°ë”ì—ì„œ í–¥í›„ ì´ë²¤íŠ¸ ê²€ìƒ‰
        
        Returns:
            ë¯¸ë˜ ì´ë²¤íŠ¸ ëª©ë¡ (í˜„ì¬ ì‹œê° ì´í›„)
        """
        events = state["calendar"].get("events", [])
        now = datetime.now(timezone.utc)
        
        # ë¯¸ë˜ ì´ë²¤íŠ¸ë§Œ í•„í„°ë§
        return [e for e in events 
                if datetime.fromisoformat(e["start"]["dateTime"]) > now]
    
    async def analyze_tasks(self, state: AcademicState) -> List[Dict]:
        """
        í•  ì¼ ëª©ë¡ ë¶„ì„
        
        Returns:
            ì „ì²´ ê³¼ì œ ëª©ë¡
        """
        return state["tasks"].get("tasks", [])
    
    async def check_learning_style(self, state: AcademicState) -> AcademicState:
        """
        í•™ìƒì˜ í•™ìŠµ ìŠ¤íƒ€ì¼ê³¼ íŒ¨í„´ ì¡°íšŒ
        
        Returns:
            í•™ìŠµ ìŠ¤íƒ€ì¼ ì •ë³´ê°€ ì¶”ê°€ëœ state
        """
        profile = state["profile"]
        
        # í•™ìŠµ ì„ í˜¸ë„ ì¶”ì¶œ
        learning_data = {
            "style": profile.get("learning_preferences", {}).get("learning_style", {}),
            "patterns": profile.get("learning_preferences", {}).get("study_patterns", {})
        }
        
        # ê²°ê³¼ë¥¼ stateì— ì €ì¥
        if "results" not in state:
            state["results"] = {}
        state["results"]["learning_analysis"] = learning_data
        
        return state
    
    async def check_performance(self, state: AcademicState) -> AcademicState:
        """
        í˜„ì¬ í•™ì—… ì„±ì·¨ë„ í™•ì¸
        
        Returns:
            ì„±ì  ì •ë³´ê°€ ì¶”ê°€ëœ state
        """
        profile = state["profile"]
        
        # ìˆ˜ê°• ì¤‘ì¸ ê³¼ëª© ì •ë³´ ì¶”ì¶œ
        courses = profile.get("academic_info", {}).get("current_courses", [])
        
        # ê²°ê³¼ë¥¼ stateì— ì €ì¥
        if "results" not in state:
            state["results"] = {}
        state["results"]["performance_analysis"] = {"courses": courses}
        
        return state

print("âœ… ReActAgent í´ë˜ìŠ¤ ì •ì˜ ì™„ë£Œ")
````

````python
# ===================================
# 1. analyze_context: í•™ìƒ ìƒí™© ë¶„ì„
# ===================================

async def analyze_context(state: AcademicState) -> Dict:
    """
    í•™ìƒì˜ í˜„ì¬ ìƒí™©ì„ ë¶„ì„í•˜ì—¬ Coordinatorì—ê²Œ ì œê³µ
    
    Returns:
        í•™ìƒ í”„ë¡œí•„, ì¼ì •, ê³¼ì œ, í•™ìŠµ ì„ í˜¸ë„ ë“±ì„ í¬í•¨í•œ ë”•ì…”ë„ˆë¦¬
    """
    profile = state.get("profile", {})
    calendar = state.get("calendar", {})
    tasks = state.get("tasks", {})
    messages = state.get("messages", [])
    
    # ìµœê·¼ ë©”ì‹œì§€ì—ì„œ í˜„ì¬ ê³¼ëª© ì¶”ì¶œ
    current_course = None
    if messages:
        last_message = messages[-1].content.lower()
        # ê°„ë‹¨í•œ í‚¤ì›Œë“œ ë§¤ì¹­ (ì‹¤ì œë¡œëŠ” ë” ì •êµí•œ ë¶„ì„ í•„ìš”)
        for course in profile.get("academic_info", {}).get("current_courses", []):
            if course.get("name", "").lower() in last_message:
                current_course = course
                break
    
    # ë¶„ì„ ê²°ê³¼ êµ¬ì¡°í™”
    context = {
        "student_info": {
            "name": profile.get("name", "Unknown"),
            "major": profile.get("academic_info", {}).get("major", "Unknown")
        },
        "learning_preferences": profile.get("learning_preferences", {}),
        "upcoming_events": len(calendar.get("events", [])),
        "active_tasks": len(tasks.get("tasks", [])),
        "current_course": current_course
    }
    
    return context

# ===================================
# 2. COORDINATOR_PROMPT: ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸
# ===================================

COORDINATOR_PROMPT = """ë‹¹ì‹ ì€ í•™ì—… ì§€ì› ì—ì´ì „íŠ¸ë“¤ì„ ì¡°ìœ¨í•˜ëŠ” Coordinatorì…ë‹ˆë‹¤.

ì‚¬ìš© ê°€ëŠ¥í•œ ì—ì´ì „íŠ¸:
â€¢ PLANNER: ì¼ì • ê´€ë¦¬ ë° ì‹œê°„ ë°°ë¶„
â€¢ NOTEWRITER: í•™ìŠµ ìë£Œ ì‘ì„± ë° ìš”ì•½
â€¢ ADVISOR: ê°œì¸í™”ëœ í•™ìŠµ ì¡°ì–¸

ë³‘ë ¬ ì‹¤í–‰ ê·œì¹™:
1. ë…ë¦½ì ì¸ ì—ì´ì „íŠ¸ëŠ” ë™ì‹œ ì‹¤í–‰ ê°€ëŠ¥
2. ì˜ì¡´ì„± ìˆëŠ” ì‘ì—…ì€ ìˆœì°¨ ì‹¤í–‰
3. ê²°ê³¼ë¥¼ í†µí•©í•˜ì—¬ ì‚¬ìš©ìì—ê²Œ ì œê³µ

ReACT íŒ¨í„´:
Thought: [ìš”ì²­ ë¶„ì„ ë° í•„ìš”í•œ ì§€ì› ìœ í˜• íŒŒì•…]
Action: [ìµœì ì˜ ì—ì´ì „íŠ¸ ì¡°í•© ì„ íƒ]
Observation: [ì„ íƒí•œ ì—ì´ì „íŠ¸ì˜ ì—­ëŸ‰ í‰ê°€]
Decision: [ìµœì¢… ì—ì´ì „íŠ¸ ë°°ì¹˜ ê³„íš]

ë¶„ì„ ìš”ì†Œ:
1. ì‘ì—… ë³µì¡ë„ì™€ ë²”ìœ„
2. ì‹œê°„ ì œì•½
3. í•„ìš” ìì›
4. í•™ìŠµ ìŠ¤íƒ€ì¼ ì í•©ì„±
5. ì§€ì› ìœ í˜•

ì»¨í…ìŠ¤íŠ¸:
ìš”ì²­: {request}
í•™ìƒ ì •ë³´: {context}

ì‘ë‹µ í˜•ì‹:
Thought: [í•™ì—… ìš”êµ¬ì‚¬í•­ ë° ìƒí™© ë¶„ì„]
Action: [ì—ì´ì „íŠ¸ ì„ íƒ ë° ê·¸ë£¹í™” ì „ëµ]
Observation: [ì˜ˆìƒ ì›Œí¬í”Œë¡œìš° ë° ì˜ì¡´ì„±]
Decision: [ìµœì¢… ì—ì´ì „íŠ¸ ë°°ì¹˜ ê³„íš ë° ê·¼ê±°]
"""

# ===================================
# 3. coordinator_agent: ë©”ì¸ ì¡°ìœ¨ í•¨ìˆ˜
# ===================================

async def coordinator_agent(state: AcademicState) -> Dict:
    """
    ReACT íŒ¨í„´ìœ¼ë¡œ ì—ì´ì „íŠ¸ë¥¼ ì¡°ìœ¨í•˜ëŠ” í•µì‹¬ í•¨ìˆ˜
    
    ì²˜ë¦¬ íë¦„:
    1. í•™ìƒ ìƒí™© ë¶„ì„ (analyze_context)
    2. LLMì—ê²Œ ìš”ì²­ ë¶„ì„ ìœ„ì„
    3. ì‘ë‹µ íŒŒì‹± ë° êµ¬ì¡°í™”
    4. ì—ì´ì „íŠ¸ ì‹¤í–‰ ê³„íš ë°˜í™˜
    
    Returns:
        required_agents: í•„ìš”í•œ ì—ì´ì „íŠ¸ ëª©ë¡
        priority: ìš°ì„ ìˆœìœ„
        concurrent_groups: ë³‘ë ¬ ì‹¤í–‰ ê·¸ë£¹
    """
    try:
        # 1. í˜„ì¬ ìƒí™© ë¶„ì„
        context = await analyze_context(state)
        query = state["messages"][-1].content
        
        # 2. LLMì—ê²Œ ì¡°ìœ¨ ìš”ì²­
        response = await llm_client.chat.completions.create(
            model=LLMConfig.MODEL_NAME,
            messages=[
                {"role": "system", "content": COORDINATOR_PROMPT.format(
                    request=query,
                    context=json.dumps(context, ensure_ascii=False, indent=2)
                )}
            ],
            max_tokens=LLMConfig.MAX_TOKENS,
            temperature=LLMConfig.TEMPERATURE
        )
        
        llm_response = response.choices[0].message.content
        
        # 3. ì‘ë‹µ íŒŒì‹±
        analysis = parse_coordinator_response(llm_response)
        
        return {
            "results": {
                "coordinator_analysis": {
                    "required_agents": analysis.get("required_agents", ["PLANNER"]),
                    "priority": analysis.get("priority", {"PLANNER": 1}),
                    "concurrent_groups": analysis.get("concurrent_groups", [["PLANNER"]]),
                    "reasoning": llm_response
                }
            }
        }
    
    except Exception as e:
        print(f"âš ï¸  Coordinator ì˜¤ë¥˜: {e}")
        # í´ë°±: PLANNERë§Œ ì‹¤í–‰
        return {
            "results": {
                "coordinator_analysis": {
                    "required_agents": ["PLANNER"],
                    "priority": {"PLANNER": 1},
                    "concurrent_groups": [["PLANNER"]],
                    "reasoning": "ì—ëŸ¬ ë°œìƒìœ¼ë¡œ PLANNERë¡œ í´ë°±"
                }
            }
        }

# ===================================
# 4. parse_coordinator_response: ì‘ë‹µ íŒŒì‹±
# ===================================

def parse_coordinator_response(response: str) -> Dict:
    """
    LLM ì‘ë‹µì„ íŒŒì‹±í•˜ì—¬ ì‹¤í–‰ ê³„íš ì¶”ì¶œ
    
    íŒŒì‹± ë¡œì§:
    - "PLANNER" í‚¤ì›Œë“œ â†’ PLANNER ì—ì´ì „íŠ¸ ì¶”ê°€
    - "NOTEWRITER" ë˜ëŠ” "note" â†’ NOTEWRITER ì¶”ê°€
    - "ADVISOR" ë˜ëŠ” "guidance" â†’ ADVISOR ì¶”ê°€
    
    ë³‘ë ¬ ì‹¤í–‰ ê·œì¹™:
    - PLANNER + NOTEWRITER: ë™ì‹œ ì‹¤í–‰ ê°€ëŠ¥
    - ADVISOR: ë³´í†µ ìˆœì°¨ ì‹¤í–‰
    """
    try:
        # ê¸°ë³¸ ì„¤ì • (í•­ìƒ PLANNER í¬í•¨)
        analysis = {
            "required_agents": ["PLANNER"],
            "priority": {"PLANNER": 1},
            "concurrent_groups": [["PLANNER"]],
            "reasoning": response
        }
        
        # ReACT íŒ¨í„´ í™•ì¸
        if "Thought:" in response and "Decision:" in response:
            # NOTEWRITER í•„ìš” ì—¬ë¶€ í™•ì¸
            if "NOTEWRITER" in response or "note" in response.lower():
                analysis["required_agents"].append("NOTEWRITER")
                analysis["priority"]["NOTEWRITER"] = 2
                # PLANNERì™€ ë³‘ë ¬ ì‹¤í–‰
                analysis["concurrent_groups"] = [["PLANNER", "NOTEWRITER"]]
            
            # ADVISOR í•„ìš” ì—¬ë¶€ í™•ì¸
            if "ADVISOR" in response or "guidance" in response.lower():
                analysis["required_agents"].append("ADVISOR")
                analysis["priority"]["ADVISOR"] = 3
                # ADVISORëŠ” ë³´í†µ ìˆœì°¨ ì‹¤í–‰
        
        return analysis
    
    except Exception as e:
        print(f"âš ï¸  íŒŒì‹± ì˜¤ë¥˜: {str(e)}")
        # ì•ˆì „í•œ ê¸°ë³¸ê°’ ë°˜í™˜
        return {
            "required_agents": ["PLANNER"],
            "priority": {"PLANNER": 1},
            "concurrent_groups": [["PLANNER"]],
            "reasoning": "íŒŒì‹± ì˜¤ë¥˜ë¡œ ì¸í•œ í´ë°±"
        }

print("âœ… Coordinator Agent ì •ì˜ ì™„ë£Œ")
````

````python
class PlannerAgent(ReActAgent):
    """
    ì¼ì • ê³„íšì„ ìˆ˜ë¦½í•˜ëŠ” ì „ë¬¸ ì—ì´ì „íŠ¸
    
    ReActAgentë¥¼ ìƒì†ë°›ì•„ ê¸°ë³¸ tools(search_calendar ë“±)ë¥¼ ì‚¬ìš©í•˜ê³ ,
    ì „ë¬¸ í”„ë¡¬í”„íŠ¸ë¡œ LLMì„ í˜¸ì¶œí•©ë‹ˆë‹¤.
    """
    
    def __init__(self, llm):
        """ë¶€ëª¨ í´ë˜ìŠ¤ ì´ˆê¸°í™” (tools ìƒì†)"""
        super().__init__(llm)
        self.llm = llm
    
    async def __call__(self, state: AcademicState) -> Dict:
        """
        PlannerAgentì˜ ë©”ì¸ ì‹¤í–‰ ë¡œì§
        
        ì²˜ë¦¬ íë¦„:
        1. ìº˜ë¦°ë”ì™€ í•  ì¼ ëª©ë¡ ìˆ˜ì§‘ (ìƒì†ë°›ì€ tools ì‚¬ìš©)
        2. í•™ìƒ í”„ë¡œí•„ê³¼ í•™ìŠµ ìŠ¤íƒ€ì¼ ì¡°íšŒ
        3. LLMì—ê²Œ ê³„íš ìˆ˜ë¦½ ìš”ì²­
        4. ê²°ê³¼ ë°˜í™˜
        """
        try:
            # 1. ë°ì´í„° ìˆ˜ì§‘ (ìƒì†ë°›ì€ tools ì‚¬ìš©)
            upcoming_events = await self.search_calendar(state)
            tasks = await self.analyze_tasks(state)
            
            # 2. í•™ìƒ í”„ë¡œí•„ ì •ë³´
            profile = state.get("profile", {})
            learning_style = profile.get("learning_preferences", {}).get("learning_style", {})
            
            # 3. ì‚¬ìš©ì ìš”ì²­
            user_request = state["messages"][-1].content
            
            # 4. í”„ë¡¬í”„íŠ¸ êµ¬ì„±
            planner_prompt = f"""ë‹¹ì‹ ì€ í•™ìƒì˜ í•™ì—… ê³„íšì„ ìˆ˜ë¦½í•˜ëŠ” ì „ë¬¸ê°€ì…ë‹ˆë‹¤.

í•™ìƒ ìš”ì²­: {user_request}

í•™ìƒ ì •ë³´:
- í•™ìŠµ ìŠ¤íƒ€ì¼: {learning_style.get('primary_style', 'Unknown')}
- ì„ í˜¸ í•™ìŠµ ì‹œê°„: {learning_style.get('preferred_times', [])}

ë‹¤ê°€ì˜¤ëŠ” ì¼ì • ({len(upcoming_events)}ê°œ):
{json.dumps(upcoming_events[:3], ensure_ascii=False, indent=2)}  # ìµœëŒ€ 3ê°œë§Œ í‘œì‹œ

í•  ì¼ ëª©ë¡ ({len(tasks)}ê°œ):
{json.dumps(tasks[:3], ensure_ascii=False, indent=2)}  # ìµœëŒ€ 3ê°œë§Œ í‘œì‹œ

ë‹¤ìŒ í˜•ì‹ìœ¼ë¡œ ê³„íšì„ ì‘ì„±í•˜ì„¸ìš”:

Thought: [ìƒí™© ë¶„ì„]
Action: [í•„ìš”í•œ ì‘ì—… ì‹ë³„]
Observation: [í˜„ì¬ ì¼ì • ë° ê³¼ì œ ìƒíƒœ]
Plan: [êµ¬ì²´ì ì¸ í•™ìŠµ ê³„íš]

ê³„íšì— í¬í•¨í•  ë‚´ìš©:
- ìš°ì„ ìˆœìœ„ê°€ ë†’ì€ ê³¼ì œ ì‹ë³„
- í•™ìŠµ ì‹œê°„ ë°°ë¶„
- í•™ìƒì˜ í•™ìŠµ ìŠ¤íƒ€ì¼ì— ë§ëŠ” ì „ëµ
- íœ´ì‹ ì‹œê°„ í¬í•¨
"""
            
            # 5. LLM í˜¸ì¶œ
            response = await llm_client.chat.completions.create(
                model=LLMConfig.MODEL_NAME,
                messages=[
                    {"role": "system", "content": planner_prompt}
                ],
                max_tokens=LLMConfig.MAX_TOKENS,
                temperature=LLMConfig.TEMPERATURE
            )
            
            plan = response.choices[0].message.content
            
            # 6. ê²°ê³¼ ë°˜í™˜
            return {
                "plan": plan,
                "events_analyzed": len(upcoming_events),
                "tasks_analyzed": len(tasks)
            }
        
        except Exception as e:
            print(f"âš ï¸  PlannerAgent ì˜¤ë¥˜: {e}")
            return {
                "plan": "ê³„íš ìˆ˜ë¦½ ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. ë‚˜ì¤‘ì— ë‹¤ì‹œ ì‹œë„í•´ì£¼ì„¸ìš”.",
                "error": str(e)
            }

# ê°„ë‹¨í•œ NoteWriterAgentì™€ AdvisorAgent ìŠ¤í… (ì‹¤ì œë¡œëŠ” ë” ë³µì¡)
class NoteWriterAgent(ReActAgent):
    """í•™ìŠµ ìë£Œ ì‘ì„± ì—ì´ì „íŠ¸ (PlannerAgentì™€ ìœ ì‚¬í•œ êµ¬ì¡°)"""
    async def __call__(self, state: AcademicState) -> Dict:
        return {"note": "í•™ìŠµ ë…¸íŠ¸ê°€ ì—¬ê¸°ì— ìƒì„±ë©ë‹ˆë‹¤."}

class AdvisorAgent(ReActAgent):
    """í•™ìŠµ ì¡°ì–¸ ì œê³µ ì—ì´ì „íŠ¸ (PlannerAgentì™€ ìœ ì‚¬í•œ êµ¬ì¡°)"""
    async def __call__(self, state: AcademicState) -> Dict:
        return {"advice": "í•™ìŠµ ì¡°ì–¸ì´ ì—¬ê¸°ì— ìƒì„±ë©ë‹ˆë‹¤."}

print("âœ… ì „ë¬¸ ì—ì´ì „íŠ¸ í´ë˜ìŠ¤ ì •ì˜ ì™„ë£Œ")
````

````python
# ===================================
# ìµœì¢… ì‘ë‹µ ìƒì„± ë…¸ë“œ
# ===================================

async def response_generator(state: AcademicState) -> Dict:
    """
    ì—ì´ì „íŠ¸ë“¤ì˜ ê²°ê³¼ë¥¼ í†µí•©í•˜ì—¬ ìµœì¢… ì‚¬ìš©ì ì‘ë‹µ ìƒì„±
    """
    try:
        agent_outputs = state["results"].get("agent_outputs", {})
        
        # ê° ì—ì´ì „íŠ¸ì˜ ê²°ê³¼ë¥¼ í¬ë§·íŒ…
        response_parts = []
        
        if "planner" in agent_outputs:
            plan = agent_outputs["planner"].get("plan", "")
            response_parts.append(f"## ğŸ“… í•™ìŠµ ê³„íš\n\n{plan}")
        
        if "notewriter" in agent_outputs:
            note = agent_outputs["notewriter"].get("note", "")
            response_parts.append(f"\n\n## ğŸ“ í•™ìŠµ ë…¸íŠ¸\n\n{note}")
        
        if "advisor" in agent_outputs:
            advice = agent_outputs["advisor"].get("advice", "")
            response_parts.append(f"\n\n## ğŸ’¡ í•™ìŠµ ì¡°ì–¸\n\n{advice}")
        
        final_response = "\n".join(response_parts) if response_parts else "ì‘ë‹µì„ ìƒì„±í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
        
        return {"final_response": final_response}
    
    except Exception as e:
        return {"final_response": f"ì‘ë‹µ ìƒì„± ì¤‘ ì˜¤ë¥˜: {str(e)}"}

# ===================================
# LangGraph ì›Œí¬í”Œë¡œìš° êµ¬ì„±
# ===================================

def create_atlas_workflow() -> StateGraph:
    """
    ATLAS ì‹œìŠ¤í…œì˜ ì „ì²´ ì›Œí¬í”Œë¡œìš° ê·¸ë˜í”„ ìƒì„±
    
    Returns:
        ì»´íŒŒì¼ëœ StateGraph (ì‹¤í–‰ ê°€ëŠ¥í•œ ì›Œí¬í”Œë¡œìš°)
    """
    # 1. StateGraph ì´ˆê¸°í™”
    workflow = StateGraph(AcademicState)
    
    # 2. AgentExecutor ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
    executor = AgentExecutor(llm_client)
    
    # 3. ë…¸ë“œ ì¶”ê°€
    workflow.add_node("coordinator", coordinator_agent)
    workflow.add_node("executor", executor.execute)
    workflow.add_node("response_generator", response_generator)
    
    # 4. ì—£ì§€ ì¶”ê°€ (ì‹¤í–‰ ìˆœì„œ ì •ì˜)
    workflow.add_edge("coordinator", "executor")
    workflow.add_edge("executor", "response_generator")
    
    # 5. ì‹œì‘ì  ì„¤ì •
    workflow.set_entry_point("coordinator")
    
    # 6. ì¢…ë£Œì  ì„¤ì •
    workflow.add_edge("response_generator", END)
    
    # 7. ì»´íŒŒì¼ (ì‹¤í–‰ ê°€ëŠ¥í•˜ê²Œ ë³€í™˜)
    return workflow.compile()

# ì›Œí¬í”Œë¡œìš° ìƒì„±
atlas_workflow = create_atlas_workflow()

print("âœ… ATLAS ì›Œí¬í”Œë¡œìš° ìƒì„± ì™„ë£Œ")
print("ğŸ“Š ë…¸ë“œ: coordinator â†’ executor â†’ response_generator â†’ END")
````

## StateGraphì˜ ì—­í• :
- **ë…¸ë“œ**: ê° ì²˜ë¦¬ ë‹¨ê³„ (coordinator, executor, response)
- **ì—£ì§€**: ë…¸ë“œ ê°„ ì—°ê²° (ì‹¤í–‰ ìˆœì„œ)
- **State**: ëª¨ë“  ë…¸ë“œê°€ ê³µìœ í•˜ëŠ” ë°ì´í„° (AcademicState)

## ì¥ì :
- ëª…í™•í•œ ë°ì´í„° íë¦„
- ê° ë‹¨ê³„ ë…ë¦½ì ìœ¼ë¡œ í…ŒìŠ¤íŠ¸ ê°€ëŠ¥

````python
class AgentExecutor:
    """
    ì—¬ëŸ¬ ì—ì´ì „íŠ¸ë¥¼ ë³‘ë ¬/ìˆœì°¨ë¡œ ì‹¤í–‰í•˜ëŠ” ì‹¤í–‰ê¸°
    
    Coordinatorì˜ ë¶„ì„ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì—ì´ì „íŠ¸ë“¤ì„ ì¡°ìœ¨í•©ë‹ˆë‹¤.
    """
    
    def __init__(self, llm):
        """ì „ë¬¸ ì—ì´ì „íŠ¸ë“¤ ì´ˆê¸°í™”"""
        self.llm = llm
        self.agents = {
            "PLANNER": PlannerAgent(llm),
            "NOTEWRITER": NoteWriterAgent(llm),
            "ADVISOR": AdvisorAgent(llm)
        }
    
    async def execute(self, state: AcademicState) -> Dict:
        """
        Coordinator ë¶„ì„ì„ ë°”íƒ•ìœ¼ë¡œ ì—ì´ì „íŠ¸ ì‹¤í–‰
        
        ì²˜ë¦¬ íë¦„:
        1. Coordinator ë¶„ì„ ê²°ê³¼ ì½ê¸°
        2. concurrent_groupsë³„ë¡œ ë³‘ë ¬ ì‹¤í–‰
        3. ê²°ê³¼ ìˆ˜ì§‘ ë° ë°˜í™˜
        """
        try:
            # 1. Coordinator ë¶„ì„ ê²°ê³¼ ì¶”ì¶œ
            analysis = state["results"].get("coordinator_analysis", {})
            required_agents = analysis.get("required_agents", ["PLANNER"])
            concurrent_groups = analysis.get("concurrent_groups", [["PLANNER"]])
            
            print(f"ğŸ“‹ ì‹¤í–‰í•  ì—ì´ì „íŠ¸: {required_agents}")
            print(f"ğŸ”„ ë³‘ë ¬ ê·¸ë£¹: {concurrent_groups}")
            
            # 2. ê²°ê³¼ ì €ì¥ì†Œ
            results = {}
            
            # 3. ê° ê·¸ë£¹ë³„ë¡œ ìˆœì°¨ ì‹¤í–‰ (ê·¸ë£¹ ë‚´ì—ì„œëŠ” ë³‘ë ¬)
            for group in concurrent_groups:
                # ê·¸ë£¹ ë‚´ ì—ì´ì „íŠ¸ë“¤ì˜ íƒœìŠ¤í¬ ì¤€ë¹„
                tasks = []
                for agent_name in group:
                    if agent_name in required_agents and agent_name in self.agents:
                        tasks.append(self.agents[agent_name](state))
                
                # ë³‘ë ¬ ì‹¤í–‰ (asyncio.gather)
                if tasks:
                    print(f"âš¡ {group} ì‹¤í–‰ ì¤‘...")
                    group_results = await asyncio.gather(*tasks, return_exceptions=True)
                    
                    # ì„±ê³µí•œ ê²°ê³¼ë§Œ ì €ì¥
                    for agent_name, result in zip(group, group_results):
                        if not isinstance(result, Exception):
                            results[agent_name.lower()] = result
                            print(f"  âœ… {agent_name} ì™„ë£Œ")
                        else:
                            print(f"  âš ï¸  {agent_name} ì˜¤ë¥˜: {result}")
            
            # 4. í´ë°±: ê²°ê³¼ê°€ ì—†ìœ¼ë©´ PLANNER ì‹¤í–‰
            if not results and "PLANNER" in self.agents:
                print("âš ï¸  í´ë°±: PLANNER ì‹¤í–‰")
                planner_result = await self.agents["PLANNER"](state)
                results["planner"] = planner_result
            
            # 5. ê²°ê³¼ ë°˜í™˜
            return {
                "results": {
                    "agent_outputs": results
                }
            }
        
        except Exception as e:
            print(f"âš ï¸  AgentExecutor ì˜¤ë¥˜: {e}")
            return {
                "results": {
                    "agent_outputs": {
                        "planner": {
                            "plan": "ì‹œìŠ¤í…œ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•´ì£¼ì„¸ìš”."
                        }
                    }
                }
            }

print("âœ… AgentExecutor í´ë˜ìŠ¤ ì •ì˜ ì™„ë£Œ")
````
